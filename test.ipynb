```python
import matplotlib.pyplot as plt
```

```python
model_counts = c4mip_data['source_id'].value_counts()
plt.figure(figsize=(10, 6))
model_counts.plot(kind='bar')
plt.title('Number of C4MIP Datasets by Model')
plt.xlabel('Climate Model')
plt.ylabel('Number of Datasets')
plt.tight_layout()
plt.show()
```

```python
exp_counts = c4mip_data['experiment_id'].value_counts()
plt.figure(figsize=(8, 6))
exp_counts.plot(kind='bar')
plt.title('C4MIP Experiment Types Available')
plt.show()
```

```python
# For question: "How does carbon cycle feedback amplify warming in different regions?"
# Step 1: Refine Your Data Filtering
# Focus on the key experiments for comparing carbon feedback
experiments = ['1pctCO2-bgc', '1pctCO2-rad']

# Filter for temperature data in these experiments
tas_feedback = c4mip_data[
    (c4mip_data['variable_id'] == 'tas') & 
    (c4mip_data['experiment_id'].isin(experiments))
]

print(f"Found {len(tas_feedback)} temperature datasets")
print(f"\nModels available: {tas_feedback['source_id'].unique()}")
print(f"\nDatasets per experiment:")
print(tas_feedback['experiment_id'].value_counts())
```

```python
# Step 2: Create Exploratory Graph - Model Availability
import seaborn as sns

# Which models have BOTH experiments?
model_exp = tas_feedback.groupby(['source_id', 'experiment_id']).size().unstack(fill_value=0)

# Only keep models that have both experiments
complete_models = model_exp[(model_exp['1pctCO2-bgc'] > 0) & (model_exp['1pctCO2-rad'] > 0)]

plt.figure(figsize=(10, 6))
complete_models.plot(kind='bar', color=['#e74c3c', '#3498db'])
plt.title('Models with Both Carbon Feedback Experiments', fontsize=14, fontweight='bold')
plt.xlabel('Climate Model', fontsize=12)
plt.ylabel('Number of Datasets', fontsize=12)
plt.legend(['With Carbon Feedback (bgc)', 'Without Carbon Feedback (rad)'], loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n{len(complete_models)} models have both experiments for proper comparison")
```

```python
#Step 3: Create Regional Analysis Preview
# Show which regions we'll analyze
regions = {
    'Amazon Rainforest': 'High carbon storage, drought sensitivity',
    'Arctic': 'Permafrost carbon release',
    'Tropical Oceans': 'Ocean carbon sink weakening',
    'Northern Forests': 'Boreal forest expansion',
    'Sahel': 'Vegetation-rainfall feedbacks'
}

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('tight')
ax.axis('off')

table_data = [[region, description] for region, description in regions.items()]
table = ax.table(cellText=table_data, 
                colLabels=['Region of Interest', 'Expected Carbon Feedback'],
                cellLoc='left',
                colWidths=[0.3, 0.7],
                loc='center')
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2.5)

# Style header
for i in range(2):
    table[(0, i)].set_facecolor('#3498db')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Alternate row colors
for i in range(1, len(regions) + 1):
    if i % 2 == 0:
        for j in range(2):
            table[(i, j)].set_facecolor('#ecf0f1')

plt.title('Key Regions for Carbon Feedback Analysis', fontsize=14, fontweight='bold', pad=20)
plt.show()
```
