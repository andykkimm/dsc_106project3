```python
import pandas as pd
```

```python
# The huge catalog
df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
df
```

```python
# Catalog with only C4MIP
c4mip_data = df[df.activity_id == 'C4MIP']

# All the different climate variables that exist in the C4MIP dataset
print(c4mip_data.variable_id.unique())
```

```python
c4mip_data[c4mip_data['variable_id'] == 'tas']
```

```python
row = c4mip_data.iloc[0]
zstore = row['zstore']
```

```python
import fsspec
import xarray as xr
```

```python
import matplotlib.pyplot as plt
```

```python
model_counts = c4mip_data['source_id'].value_counts()
plt.figure(figsize=(10, 6))
model_counts.plot(kind='bar')
plt.title('Number of C4MIP Datasets by Model')
plt.xlabel('Climate Model')
plt.ylabel('Number of Datasets')
plt.tight_layout()
plt.show()
```

```python
exp_counts = c4mip_data['experiment_id'].value_counts()
plt.figure(figsize=(8, 6))
exp_counts.plot(kind='bar')
plt.title('C4MIP Experiment Types Available')
plt.show()
```
